## 3-1 언패킹

```python
coordinates = (10, 20)

# 튜플 언패킹
x,y = coordinates
print(f'x좌표: {x}, y좌표: {y}')

rgb = [255, 128, 0]
# 리스트 언패킹
red, green, blue = rgb
print(f'빨강: {red}, 초록: {green}, 파랑: {blue}')

word = 'ABC'
# 문자열 언패킹
first, second, third = word
print(f'첫 번째 문자: {first}, 두 번째 문자: {second}, 세 번째 문자: {third}')
```

```python
def get_dataset_stats(values):
  return min(values), max(values), sum(values) / len(values)

data = [12, 8, 21, 17, 5, 32, 14]

# 함수값 return 언패킹
minimum, maximum, average = get_dataset_stats(data)
print(f'최솟값: {minimum}, 최댓값: {maximum}, 평균: {average:.2f}')
```

```python
student_scores = [
    ('김철수', 85, 92, 78),
    ('이영희', 92, 88, 95),
    ('박지민', 75, 83, 90)
]

# 반복문에서 언패킹 사용
for name, database, python, cloud in student_scores:
    average = (database + python + cloud) / 3
    print(f'{name}의 평균 점수: {average:.1f}')
```

```python
monthly_sales = [1200, 1350, 1420, 1500, 1300, 1580, 1620, 1700, 1800, 1850, 1900, 2000]

# 1월, 2월 및 나머지 월 매출액
first, second, *remaining = monthly_sales
print(f'1월 판매액: {first}')
print(f'2월 판매액: {second}')
print(f'나머지 월 판매액: {remaining}')

# 1월, 12월 및 나머지 월 매출액
first, *middle, last = monthly_sales
print(f'첫 달(1월) 판매액: {first}')
print(f'마지막 달(12월) 판매액: {last}')
print(f'중간 달(2~11월)의 판매액: {middle}')
```

## 3-3 함수형 프로그래밍

### 람다 함수


*   이름이 없는 익명함수
>한 줄의 표현식으로 작성할 수 있는 간결한 함수  
> 즉시 사용할 수 있는 간결한 연상르 수행할 때 적합  
>> *lambda 매개변수 : 표현식*

*   함다 함수의 문법적 특징
> 단 하나의 표현식만 허용  
> 표현식의 결과는 자동으로 함수의 반환값이 되므로 return 키워드를 사용하지 않음



```python
def add(x):
    return x+2

add_lambda = lambda x: x+2

print(add(3))
print(add_lambda(3))
```

```python
concat = lambda s1, s2 : s1+' '+s2   # 두 개의 문자열을 받아 공백을 추가후 결합
print(concat('파이썬','데이터 분석'))

is_even = lambda x: '짝수' if x % 2 == 0 else '홀수'

print(is_even(10))
print(is_even(7))
```

```python
employees = {
    '이지혜': 3000000,
    '구민준': 5000000,
    '방서연': 4100000
}

# 함수 정의: salary * 1.1
adjust_salary = lambda salary: salary * 1.1
# 급여 인상
updated_salaries = {name : adjust_salary(salary) for name, salary in employees.items()}
print(updated_salaries)
```

#### map 함수 : 반복 가능한 객체의 모든 요소에 특정 함수를 적용하여 새로운 값을 생성하는 함수
- 리스트나 튜플과 같은 자료구조의 요소를 변경할 때 유용

```python
temperature_celsius = [25.6, 27.8, 30.5, 22.3, 28.9, 31.2, 24.7]

# 함수 정의: 섭씨 -> 화씨
celsius_to_fahrenheit = lambda c :(c *9/5) +32

# 온도 변환
temperature_fahrenheit = list(map(celsius_to_fahrenheit, temperature_celsius))

print('섭씨 온도 데이터:', temperature_celsius)
print('화씨 온도로 변환:', temperature_fahrenheit)
```

```python
words = ['python', 'data', 'analysis']

# 함수 정의 및 대문자화
uppercase_words = list(map(lambda word: word.upper(), words))
print(uppercase_words)
```

```python
employees = [
    {'name': '김지원', 'salary': 3700000},
    {'name': '박민준', 'salary': 7000000},
    {'name': '이서연', 'salary': 4400000},
    {'name': '정현우', 'salary': 3500000}
]

# salary * 1.1
updated_employees = list(map(lambda emp : {'name': emp['name'],
                                           'salary': emp['salary']* 1.1}, employees))
print('\n급여 인상 후:')

for emp in updated_employees:
    print(f'{emp['name']}: {int(emp['salary']):,}원')
```

#### filter 함수  
* 반복가능한 객체의 요소 중 특정 조건을 만족하는 요소  
> True 또는 False를 반환하는 함수(조건함수)를 인자로 받고 조건함수가 True를 반환하는 요소만 컬렉션에 포함

```python
numbers = [10, 15, 20, 25, 30]

# 짝수 필터링
even_numbers = list(filter(lambda x : x %2 ==0, numbers))
print(even_numbers)
```

```python
osda_students = [
    {'name': '김지원', 'score': 85},
    {'name': '이민준', 'score': 65},
    {'name': '박서연', 'score': 90},
    {'name': '정현우', 'score': 55},
    {'name': '최예은', 'score': 78}
]

# 70점 이상 학생 목록 생성
passed_students = list(filter(lambda student : student['score']>=70,
                              osda_students))

for student in passed_students:
    print(f'이름: {student["name"]}, 점수: {student["score"]}')
```

#### reduce 함수
* 반복 가능한 객체의 모든 요소를 누적하여 하나의 값으로 축소하는 함수  
> 반복문을 사용하지 않고도 데이터를 하나의 값으로 집계  
> functools 모듈에서 제공
>>1. 첫번째와 두번째 요소를 함수에 적용하여 중간 결과를 생성
>>2. 그 중간 결과와 객체의 다음 요소를 다시 함수에 적용  
>>3. 이 과정을 반복하여 최종적으로 하나의 값을 반환

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total_sum = reduce(lambda x, y : x+y, numbers)
print(total_sum)
```

```python
from functools import reduce

numbers = [3, 7, 2, 9, 5]

# 최대값 생성
max_value = reduce(lambda x, y : x if x > y else y, numbers)
print(max_value)
```
